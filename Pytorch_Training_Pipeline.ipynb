## Convert numpy arrays into tensors

```python
import torch

X_train_tensor = torch.from_numpy(X_train_scaled)

y_train_tensor = torch.from_numpy(y_train_n)

X_test_tensor = torch.from_numpy(X_Test_scaled)

y_test_tensor = torch.from_numpy(y_test_n)
```

```python
X_train_tensor.shape
```

## creating the neural network

```python
class simpleNN:
  def __init__(self,input_dim,criteria):
    self.weights = torch.randn(input_dim,1,dtype = torch.float32,requires_grad=True)      # initialize weight matrix
    self.bias = torch.zeros(1 ,dtype = torch.float32 ,requires_grad=True)                 # initialize bias with 0
    self.criteria = criteria                       # critetria = loss func = binary cross entropy
    self.lr = 0.1                                  # learning rate


  def forwardpass(self , X):
    z = torch.matmul(X , self.weights) + self.bias      # calcuate weighted sum
    y_pred = torch.sigmoid(z)                           # pass it to sigmoid
    return y_pred                                     # return prediction


  def backprop(self,X,y_true):
    y_pred = self.forwardpass(X)          # get the predicted value
    loss = self.criteria(y_pred,y_true)  # calculate loss
    loss.backward()                       # find partial derivative

    with torch.no_grad():                           # do not save the gradient in computation graph
      self.weights -= (self.lr)*self.weights.grad   # update weight
      self.bias -=  (self.lr)*self.bias.grad        # update bias

    self.weights.grad.zero_()        # prevent the gradients from accumulating
    self.bias.grad.zero_()

    return loss.item()
```

```python
from torch import nn
criteria = nn.BCELoss()     # binary cross entropy
model = simpleNN (X_train_tensor.shape[1],criteria)
epochs = 100
for i in range(epochs):
  loss = model.backprop(X_train_tensor.float() , y_train_tensor.float().unsqueeze(1))
  print(f"Epoch {i} , loss{loss}")
```

## Results

```python
from sklearn.metrics import accuracy_score
with torch.no_grad():
  y_pred = model.forwardpass(X_test_tensor.float())
  y_pred =(y_pred > 0.5).int() # threshold

acc = accuracy_score(y_test ,y_pred.numpy())
print(f"Accuracy: {acc}")
```
